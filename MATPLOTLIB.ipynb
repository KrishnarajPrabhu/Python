Operator Description

```python
from matplotlib import pyplot as plt
```

```python
fig=plt.figure()

axis=fig.add_axes([0.5,0.1,0.8,0.8]) #control the left,right,width,height of the canvas (from 0 to 1)

axis.plot(x,y,'r')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=3) #Submap is of 1 row, 2 columns
for ax in axes:
    ax.plot(x,y,'r')
```

```python
fig,axes=plt.subplots(nrows=2,ncols=3) #Submap is of 1 row, 2 columns
for ax in axes:
    ax.plot(x,y,'r')
```

```python
#graph inside another graph
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8]) #big axes
axes2=fig.add_axes([0.2,0.5,0.4,0.3]) #small canvas
axes1.plot(x,y,'r')
axes2.plot(y,x,'g')
```

```python
#using colour ,transparency, etc
fig,axes=plt.subplots(dpi=150)
axes.plot(x,x+1,color='red')
axes.plot(x,x+2, color='blue')
axes.plot(x,x+3, color='green')
```

```python
fig,axes=plt.subplots(dpi=150)
axes.plot(x,x**2,color='red',alpha=.8)
axes.plot(x,x+2, color='blue',alpha=.8)
axes.plot(x,x+3, color='green',alpha=.8)
```

```python
#using line width
fig,ax=plt.subplots(dpi=100)
ax.plot(x,x+1,color='blue',linewidth=0.25)
ax.plot(x,x+2,color='blue',linewidth=1)
ax.plot(x,x+3,color='blue',linewidth=1.5)
ax.plot(x,x+4,color='blue',linewidth=2)
```

```python
fig,ax=plt.subplots(dpi=100)
ax.plot(x,x+1,color='blue',linestyle='-')
ax.plot(x,x+2,color='blue',linestyle='-.')
ax.plot(x,x+3,color='blue',linestyle=':')
```

```python
fig, ax = plt.subplots(dpi=100)
ax.plot(x, x+1, color='blue', linestyle='-')
ax.plot(x, x+2, color='blue', linestyle='-.')
ax.plot(x, x+3, color='blue', linestyle=':')

ax.plot(x, x+4, color='black', lw=1.50, dashes=[5, 10, 15, 10])
ax.plot(x, x+4, color='black', lw=1.50, dashes=[5, 10, 15, 3])
ax.plot(x, x+4, color='black', lw=1.50, dashes=[5, 10, 4, 10])
```

```python
n=np.array([0,1,2,3,4,5])
fig,axes=plt.subplots(1,4,figsize=(16,5))

axes[0].set_title('scatter')
axes[0].scatter(x,x+0.25*np.random.randn(len(x)))

axes[1].set_title('step')
axes[1].step(n,n**2,lw=2)

axes[2].set_title('bar')
axes[2].bar(n, n**2, align='center', alpha=0.5)

axes[3].set_title("fill_between")
axes[3].fill_between(x, x**2, x**3, color='green', alpha=0.5)
```

```python
#draw contour image
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
```

```python
delta=0.999
x=np.arange(-3.0,3.0,delta)
y=np.arange(-2.0,2.0,delta)
X,Y=np.meshgrid(x,y)
Z1=np.exp(-X**2-Y**1)
Z2=np.exp(-(X-1)**2-(Y-1)**2)
Z=(Z1-Z2)*2
```

```python
fig,ax=plt.subplots()
CS=ax.contour(X,Y,Z)
ax.clabel(CS,inline=1,fontsize=10)
ax.set_title('contour map')
```
